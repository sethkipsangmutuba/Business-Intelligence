## 3.2 Data Warehousing: Definitions and Concepts

##Seth Kipsang

### Introduction
Data warehousing (DW) has become a **cornerstone of managerial decision support**, enabling organizations to **consolidate, analyze, and act** on information from diverse operational systems. When integrated with **Decision Support Systems (DSS)** and **Business Intelligence (BI)** tools, modern DWs support **real-time decision-making** by providing a **unified, reliable, and analytically-ready view** of enterprise data. The Isle of Capri EDW case demonstrates how **active data warehouses** transform fragmented operational data into actionable insights, enhancing **business processes, customer understanding, and operational agility**.

---

### What Is a Data Warehouse?
A **data warehouse** is a **centralized repository** of integrated, historical, and current data, specifically designed to support **decision-making and analytics**.  

**Key Attributes:**
- **Subject-Oriented:** Organized around core business subjects (e.g., customers, products, sales) for detailed performance analysis.  
- **Integrated:** Consolidates heterogeneous sources into a consistent format, resolving conflicts in naming conventions, units, and formats.  
- **Time-Variant:** Stores historical data for trend analysis, forecasting, and long-term comparisons.  
- **Nonvolatile:** Data are read-only for analysis; updates generate new records, preserving historical integrity.  

**Additional Characteristics:**
- **Web-enabled:** Accessible via browser interfaces.  
- **Relational or Multidimensional:** Optimized for **OLAP**, querying, and advanced analytics.  
- **Client/Server Architecture:** Facilitates organization-wide access.  
- **Real-Time / Active DW:** Supports immediate analysis and operational decision-making.  
- **Metadata-Driven:** Contains data about data to guide interpretation and usage.

---

### Historical Perspective
- **Early 1900s–1970s:** Manual aggregation for trend analysis; mainframe computing with complex files limited timely access.  
- **Late 1970s:** Teradata develops parallel-processing DBMS for decision support, laying the foundation for large-scale DW.  
- **1980s:** PCs, distributed DBMS, and relational databases emerge; fragmented “islands of data” highlight need for integration.  
- **1990s:** Bill Inmon and Ralph Kimball formalize DW concepts, emphasizing centralized repositories, dimensional modeling, and enterprise integration.  
- **2000s:** Consolidation of DW/BI vendors (Microsoft, Oracle, SAP, IBM); introduction of DW appliances for high-performance analytics.  
- **2010s:** Big Data technologies (Hadoop, NoSQL, MapReduce) extend DW to handle **unstructured and massive datasets**.

---

### Types of Data Warehouses

**Data Marts**  
- Subsets of a DW focused on a single business area or department.  
- **Dependent:** Derived from an EDW to ensure consistency.  
- **Independent:** Standalone, cost-effective solutions for small units or strategic business units.

**Operational Data Stores (ODS)**  
- Interim, near real-time repositories for current operational data.  
- Support short-term decision-making and often act as **staging areas for EDWs**.  
- Enable **oper marts** for multidimensional analysis of operational data.

**Enterprise Data Warehouses (EDW)**  
- Large-scale, enterprise-wide repositories integrating diverse sources.  
- Provide a **single version of the truth**.  
- Support DSS applications, including **CRM, SCM, BPM, BAM, and knowledge management**.

**Metadata**  
- **Definition:** Data about data, essential for effective DW usage.  
- **Types:** Technical (structure/syntax), Business (context/meaning), Semantic (domain-specific interpretation).  
- **Functions:** Enables understanding, navigation, contextualization, and informed decision-making.  
- **Implementation:** Enterprise metadata strategies may include **ontology, metadata registries, ETL processes, and SOA**.  
- **Maturity Levels (Zhao, 2005):** Ad hoc → Discovered → Managed → Optimized → Fully automated.  
- **Ethical Considerations:** Privacy, ownership, and intellectual property concerns arise in metadata collection and usage.

---

### Key Takeaways
- **Data warehouses** are repositories; **data warehousing** is the process of designing, building, and managing them to support decisions.  
- Integration of **operational, historical, and external data** enables robust analytics and BI.  
- **Metadata** is critical for contextual understanding, knowledge creation, and effective utilization of warehouse data.  
- Choosing between **data marts, ODS, or EDW** depends on scope, cost, and decision-making needs.

## 3.3 Data Warehousing Process Overview

### Introduction
Modern organizations—public and private—collect **data, information, and knowledge** at an exponentially increasing rate. The challenge lies not only in storing datasets but also in **organizing, integrating, and making them accessible** for timely decision-making. As **users and networked systems** proliferate, scalability and accessibility issues intensify. Working across multiple databases, whether integrated into a **data warehouse (DW)** or siloed, requires specialized expertise but offers **substantial business value**.

**Case Example:** Teradata’s EDW implementation for a major automobile manufacturer demonstrates that an **enterprise-wide warehouse** can reduce costs, improve quality, and accelerate business analytics.

---

### The Role of Data Warehousing in Decision Support
Data warehouses consolidate **historical, operational, and external data** into a single, coherent repository. By providing a **single version of the truth**, they reduce fragmentation, improve data reliability, and enable **analytic decision support**.

**Healthcare Example: MultiCare Health System**
- MultiCare deployed **Health Catalyst’s Adaptive Data Warehouse** to integrate clinical, financial, and operational data.  
- Resulted in rapid identification of **sepsis patients**, reducing septicemia mortality by **22% in one year** and saving over **$1.3 million**.  
- Cross-disciplinary teams (clinicians, IT, analysts) leveraged the DW to **standardize care protocols, monitor patient trends, and implement process improvements**.  

**Key Insight:** A DW is not merely a repository; it is a **strategic enabler** for improving outcomes, optimizing operations, and supporting mission-critical decisions.

---

### The Data Warehousing Process
The DW process involves several systematic steps:

**1. Data Sources**
- **Internal:** ERP, OLTP systems, legacy operational databases.  
- **External:** Census data, market research, web logs, IoT devices.  
- Data can be **structured, semi-structured, or unstructured**.

**2. Data Extraction and Transformation (ETL)**
- Extract data from multiple sources.  
- Transform to ensure **consistency, standardization, and cleansing**.  
- ETL tools may be **custom-built** or **commercial solutions**.

**3. Data Loading**
- Data loaded into a **staging area** for final transformation and cleansing.  
- Subsequently moved into the **DW** or **data marts**.

**4. Comprehensive Database / Enterprise Data Warehouse (EDW)**
- Central repository integrating diverse datasets.  
- Supports **queries, OLAP, reporting, and advanced analytics**.  
- Provides a **single version of truth** for the organization.

**5. Metadata Management**
- Metadata describe **structure, source, transformation rules, and usage**.  
- Facilitates **data governance, accessibility, and understanding** for analysts and end users.

**6. Middleware and Front-End Tools**
- Middleware provides **user access** to the DW.  
- Business users leverage **BI tools, dashboards, OLAP, data mining, and visualization platforms**.  
- Ensures **user-friendly interaction** without requiring deep technical expertise.

---

### Conceptual Framework

Source Systems → ETL → Staging Area → EDW → Data Marts (optional) → BI/Analytics Tools → Users

- Users can access the DW through **middleware or front-end applications** for operational, tactical, or strategic decisions.  
- Flexibility: Some organizations **bypass data marts** and allow direct EDW access using BI tools for **ad hoc queries**.

---

### Key Lessons
- **Integration:** DWs consolidate multiple sources for comprehensive decision support.  
- **ETL Processes:** Ensure **accuracy, consistency, and usability** of data.  
- **Metadata:** Essential for contextual understanding and **efficient data utilization**.  
- **Middleware & Front-End Tools:** Democratize access, empowering **non-technical decision-makers**.  
- **Business Impact:** Case studies like MultiCare show DWs can drive **measurable improvements** in performance, outcomes, and cost-efficiency.

## 3.4 Data Warehousing Architectures

### Introduction
Data warehousing architectures define the **structural organization** of data storage, processing, and access layers within an enterprise. Architecture determines **scalability, performance, flexibility, and maintainability**, which are critical for supporting **decision support systems (DSS)**, **business intelligence (BI)**, and **business analytics (BA)**.

Traditional DWs employ **client/server or n-tier architectures**, with two-tier and three-tier models being most prevalent. Emerging technologies—**Web access, cloud computing, distributed systems**—add nuance to architectural decisions.

---

### Three-Tier vs. Two-Tier Architectures

#### Three-Tier Architecture
Separates the DW system into **three functional layers**:

1. **Back-End (Data Acquisition)**
   - Extracts data from operational/legacy systems or external sources.
   - Performs **ETL (Extract, Transform, Load)**, consolidating and summarizing data.

2. **Middle Layer (Data Warehouse)**
   - Houses centralized repository (relational/multidimensional).
   - Supports complex queries, reporting, and ensures **single version of truth**.

3. **Front-End (Client/DSS/BI Engine)**
   - Provides access through analytical tools, dashboards, OLAP cubes, or custom applications.
   - Separation improves **resource optimization and scalability**.

**Advantages**
- Functional separation reduces **resource contention**.
- Supports **data marts** and **multidimensional analysis**.
- Suitable for **large-scale, performance-intensive environments**.

**Example:** Teradata EDW integrates diverse organizational data, supporting enterprise-level analytics and departmental data marts.

---

#### Two-Tier Architecture
- DW and DSS/BI engine reside on **the same hardware platform**.
- **Advantages:** Economical, simpler to implement, suitable for medium-scale warehouses.
- **Limitations:** Less scalable, potential **resource conflicts** under high query load.

---

### Web-Based Data Warehousing
- **Client:** Web browser (Java-enabled recommended) on PC/mobile.
- **Web Server:** Manages requests and interfaces with the warehouse.
- **Application Server + DW:** Processes queries, returns analytic results.

**Benefits**
- Platform independence and broad accessibility.
- Reduced deployment costs vs. traditional client/server.
- Real-time or near-real-time enterprise data access.

**Examples**
- Vanguard Group: Web-based three-tier DW for consistent data views.  
- Hilton: Migrated client/server systems to Web-based three-tier architecture; **processing efficiency increased sixfold**, expected annual savings of $4.5–5 million.

**Design Consideration:** Server capacity and page-loading speed are critical for large enterprise datasets.

---

### Alternative Data Warehousing Architectures

1. **Independent Data Marts**
   - Departmental stores developed in isolation.
   - **Pros:** Low cost, fast deployment.
   - **Cons:** Lack enterprise consistency; difficult to achieve **single version of truth**.

2. **Data Mart Bus Architecture**
   - Interconnected data marts with **conformed dimensions**.
   - Supports enterprise-level queries while maintaining departmental autonomy.
   - **Limitations:** Performance may degrade for complex cross-mart queries.

3. **Hub-and-Spoke (Corporate Information Factory)**
   - Centralized EDW (**hub**) with dependent **subject-area data marts** (**spokes**).
   - Scalable, maintainable, allows **iterative deployment**.
   - Risks: Data redundancy and latency if poorly managed.

4. **Centralized Data Warehouse**
   - Single, large EDW without dependent data marts.
   - Provides holistic enterprise-wide view.
   - Simplifies administration but can be expensive and complex at scale.

5. **Federated Data Warehouse**
   - Integrates distributed, disparate sources via middleware.
   - Suitable for heterogeneous systems.
   - Usually **supplements** rather than replaces traditional DWs; performance and data quality may be limited.

---

### Factors Influencing Architecture Selection
(Ariyachandra & Watson, 2005, 2006)
1. Interdependence of organizational information.
2. Upper management’s information needs.
3. Urgency for a data warehouse.
4. Nature of end-user tasks.
5. Resource constraints.
6. Strategic vision for DW.
7. Compatibility with existing systems.
8. In-house IT staff capabilities.
9. Technical considerations (DBMS, ETL, storage, processing).
10. Social/political factors (user involvement, executive sponsorship).

**Insight:** Technical design alone is insufficient; **organizational context, governance, and user engagement** are equally critical.

---

### Empirical Insights on Architecture Success
(Ariyachandra & Watson, 2006b; survey of 454 organizations)
| Architecture Type             | Usage | Notes |
|-------------------------------|-------|-------|
| Hub-and-Spoke                 | 39%   | Favored for large, enterprise-wide implementations |
| Bus Architecture               | 26%   | Supports conformed dimensions |
| Centralized DW                 | 17%   | Simpler administration, costly |
| Independent Data Marts         | 12%   | Lowest success across metrics |
| Federated                      | 4%    | Short-term or supplementary solution |

**Conclusion:** Effectiveness depends on **scale, complexity, enterprise needs, and user requirements**. Hub-and-spoke and bus architectures are widely successful; independent data marts are generally least effective.

---

### Section Summary
- **Three-tier vs. two-tier:** Balance performance, scalability, and cost.
- **Web-based DW:** Enhances accessibility, platform independence, and reach.
- **Alternative architectures:** Allow flexibility, customization, and incremental deployment.
- **Architecture selection:** Must consider **technical, organizational, and social factors**.
- **Empirical evidence:** Supports hub-and-spoke and bus architectures for **enterprise-scale deployments**.

## 3.5 Data Integration and the Extraction, Transformation, and Load (ETL) Processes

### Introduction
Modern enterprises face increasing pressure from **global competition, regulatory requirements, and stakeholder scrutiny**, requiring **timely, accurate, and holistic data** from heterogeneous sources—legacy systems, ERP/CRM platforms, flat files, and real-time/unstructured streams.

Prior to data warehouses, enterprise-wide data integration was **labor-intensive**, demanding specialized database expertise. As organizations scale, integration complexity rises due to **mergers, new channels, and evolving BI requirements**.

---

### Data Integration: Conceptual Framework
**Data integration** is the systematic unification of disparate data sources to produce a **consistent, accessible, and analyzable enterprise dataset**. Key processes include:

1. **Data Access**
   - Extracts data from relational databases, ERP systems, flat files, or external feeds.

2. **Data Federation**
   - Provides a **logical, unified view** across multiple data stores.
   - Physical consolidation is optional; enables cross-source analytics.

3. **Change Capture**
   - Identifies, extracts, and propagates data modifications from source systems.
   - Supports **incremental updates** for near real-time analytics.

**Example – BP Lubricants BIGS Program:**  
Following mergers, BP used Kalido’s federated DW solution to achieve **consistent reporting**, agile adaptation to changing needs, and improved **visibility and timeliness** for decision-making.

---

### Technologies Supporting Data Integration

1. **Enterprise Application Integration (EAI)**
   - Integrates data and functionality across heterogeneous systems.
   - Uses **Service-Oriented Architecture (SOA)** and Web services.
   - Enables near real-time data acquisition for OLTP and DW systems.

2. **Enterprise Information Integration (EII)**
   - Virtual integration without physically moving data.
   - Metadata-driven views present heterogeneous sources as relational tables.
   - Often XML-based for flexible knowledge representation.

3. **Physical Data Integration**
   - Traditional approach: extract, transform, and **load data** into DW/marts.
   - Ensures **consistency, quality, and centralized business rules** application.

---

### Extraction, Transformation, and Load (ETL) Processes

#### Overview
The **ETL process** is the backbone of data warehousing:

1. **Extraction**
   - Reads data from multiple source systems.

2. **Transformation**
   - Cleanses, standardizes, and reshapes data according to **business rules** (e.g., encoding, lookups, calculations).

3. **Load**
   - Ingests transformed data into the centralized DW or data mart.

**Additional ETL Functions**
- Maintain **metadata** and **data lineage**.
- Schedule and monitor runtime processes.
- Manage errors, logs, and process statistics.

> Industry studies: ETL can consume **up to 70% of project time**, highlighting its critical importance.

---

#### Best Practices
- **Staging Areas:** Temporary storage for cleansing/transformation before final loading.
- **Centralized Metadata Repository:** Ensures consistency of transformations and business rules.
- **Tool Selection:** Consider compatibility, metadata management, standards compliance, and usability for developers and functional users.

**Example – Motorola:**  
Aggregated procurement data from 30 systems into a **global SCM data warehouse** to analyze spending patterns.

---

#### Categories of ETL Technologies (Solomon, 2005)
1. **Sophisticated:** Automated, documented, scalable.
2. **Enabler:** Supports project development, requires customization.
3. **Simple:** Basic, rule-driven transformations.
4. **Rudimentary:** Minimal automation, labor-intensive.

**Strategic Insight:**  
Extensive ETL may indicate fragmented data management. Well-managed enterprises reduce redundancy, simplify ETL, and achieve **cost efficiency** and **higher data quality**.

---

#### Integration Tools and Vendor Solutions
Prominent ETL and data integration platforms include:

- SAS Data Integration Server  
- Oracle Business Intelligence Suite  
- Microsoft SSIS  
- IBM InfoSphere  
- Informatica  
- Tibco  
- Embarcadero  

Organizations must decide whether to **build ETL in-house** or **purchase commercial tools**, balancing cost, learning curve, and maintainability.

---

### Strategic Implications
Effective ETL and integration enable:

- Accurate, enterprise-wide analytics.  
- Consistent reporting across business units.  
- Rapid adaptation to regulatory/market changes.  
- Minimal operational disruption during transformations.  

Poor ETL can cause:

- Data inconsistencies.  
- High maintenance costs.  
- Delayed insights, impairing strategic decision-making.

## 3.6 Data Warehouse Development and Strategic Applications

### a. Overview
A **data warehouse (DW)** is a socio-technical system integrating operational, transactional, and historical data across departments to support enterprise-scale decision-making. Unlike OLTP systems, DWs focus on **analytics, reporting, and multidimensional queries**. Successful DW development requires attention to **architecture, governance, user engagement, cost control, and alignment with organizational strategy**.

---

### b. Benefits of Data Warehouses

#### b.1 Direct Benefits
- **Analytical Capability:** Supports multidimensional ad hoc queries, trend analysis, forecasting, and advanced analytics.  
- **Single Version of the Truth:** Reduces redundancy, ensures enterprise-wide consistency.  
- **Timely Information:** Offloads queries from operational systems, improving response times.  
- **System Performance:** Operational systems remain unburdened by heavy reporting tasks.  
- **Simplified Data Access:** Provides consistent, user-friendly access for managers, analysts, and executives.  

#### b.2 Indirect Benefits
- **Knowledge Enhancement:** Empowers informed decision-making.  
- **Competitive Advantage:** Enables faster strategic responses to market conditions.  
- **Customer Service Improvement:** Real-time insights optimize operational response.  
- **Business Process Reengineering:** Analytics identifies inefficiencies and improvement opportunities.  

#### b.3 ROI Considerations
- **Keepers (20%):** Savings from improved decision support.  
- **Gatherers (30%):** Automated collection and dissemination efficiencies.  
- **Users (50%):** Value from decisions enabled by DW access.  

---

### c. Data Warehouse Development Approaches

#### c.1 Inmon Approach (Top-Down, EDW)
- Enterprise-wide architecture using **ERD and relational modeling**.  
- High complexity, cost, and development time.  
- Objective: Provide a comprehensive, integrated view of organizational data.  

#### c.2 Kimball Approach (Bottom-Up, Data Mart)
- Builds **department-specific data marts** using dimensional modeling.  
- Faster deployment, higher early end-user accessibility.  
- Objective: Deliver immediate business value through process-focused analytics.  

#### c.3 Comparison of Approaches

| Feature | Inmon | Kimball |
|---------|-------|---------|
| Approach | Top-down | Bottom-up |
| Scope | Enterprise-wide | Departmental/subject-focused |
| Complexity | High | Moderate |
| Deployment Time | Years | Months |
| Cost | $1M+ | $10K–$100K+ |
| End-user Access | Low early | High early |
| Data Consistency | High | Conformed dimensions |

**Insight:** Many organizations adopt a **hybrid strategy**, starting with data marts for quick value, later integrating into an EDW.

---

### d. Schema Design and Data Representation

#### d.1 Fact and Dimension Tables
- **Fact Table:** Stores quantitative measures (sales, revenue, occupancy) and links to dimension tables.  
- **Dimension Tables:** Descriptive attributes (time, product, location, customer) enabling **slicing, dicing, hierarchical aggregation**.  
- **Relationship:** One-to-many from dimension tables to fact table.

#### d.2 Schema Types
- **Star Schema:** Denormalized dimensions; optimized for fast reads and simple ad hoc queries.  
- **Snowflake Schema:** Normalized dimensions; reduces redundancy, increases join complexity; suited for hierarchical queries and storage efficiency.

---

### e. OLAP and Multidimensional Analysis

#### e.1 OLAP vs. OLTP

| Feature | OLTP | OLAP |
|---------|------|------|
| Purpose | Operational transactions | Decision support/analytics |
| Data | Normalized | Denormalized/multidimensional |
| Queries | Routine, fast | Complex, ad hoc |
| Users | Operational staff | Analysts, managers |
| Performance Focus | Insert/update efficiency | Query speed for multidimensional analysis |

#### e.2 OLAP Operations
- **Slice:** Filter on a single dimension.  
- **Dice:** Filter across multiple dimensions.  
- **Drill Down/Up:** Navigate between detailed and aggregated summaries.  
- **Roll-Up:** Aggregate along a hierarchy (daily → monthly).  
- **Pivot:** Reorient data for alternative perspectives.  

#### e.3 OLAP Variants
- **ROLAP:** Relational DBs, dynamic SQL; scalable but slower for large aggregates.  
- **MOLAP:** Precomputed cubes; fast queries, higher ETL overhead.  
- **HOLAP:** Hybrid; summary in MOLAP, details in ROLAP; balances speed and flexibility.

---

### f. Vendor and Strategic Considerations

#### f.1 Vendors
- Common DW/BI vendors: SAP Business Objects, IBM InfoSphere, Oracle, Teradata, Microsoft, Informatica, Greenplum, SAS.  
- **Selection Criteria:** Performance, scalability, integration, support, cost, security.

#### f.2 Strategic Considerations

## 3.7 Data Warehousing Implementation Issues

### a. Complexity of DW Implementation
Data warehouse (DW) implementation is a **large-scale, multi-faceted endeavor** requiring cross-functional expertise in technical, managerial, and organizational domains. Key challenges include:

- Large-scale project planning and execution.  
- Coordination across multiple departments, agencies, or business units.  
- Balancing technological, analytical, and business requirements.  
- Integrating diverse data sources and formats (structured, semi-structured, unstructured).  

**Benchmarking Tools:**
- **TDWI Data Warehousing Maturity Model:** Six stages—Prenatal, Infant, Child, Teenager, Adult, Sage—track DW evolution.  
- **Characteristics Tracked:** Scope, analytic sophistication, executive perception, stewardship, funding, technology, change management, and administration.  
- Business value generally increases with maturity.  

**Insight:** DW success is measured not only technically but also strategically (Ariyachandra & Watson, 2006; Watson et al., 1999).

---

### b. Common DW Implementation Risks
DW projects are high-risk, expensive, and resource-intensive. Key risks include:

- **Inadequate executive sponsorship:** Requires influential sponsors and a strong project champion.  
- **Unrealistic expectations:** Align early “selling” with delivery phases; overpromising leads to frustration.  
- **Political naivety:** Avoid implying previous managerial decisions were poor; frame DW as decision-enabling.  
- **Data overloading:** Loading indiscriminate data slows performance; focus on curated, relevant data.  
- **Misunderstanding DW design:** DWs are **multidimensional, denormalized, and aggregate-focused**, unlike transactional OLTP systems.  
- **Technology-centric management:** Ensure **business needs drive design**, not only technology.  
- **Ignoring external/unstructured data:** Images, text, video, and external feeds can add strategic value.  
- **Conflicting data definitions:** Requires robust cleansing and reconciliation.  
- **Overestimating performance/capacity:** Plan scalable architecture for future growth.  
- **Treating DW as static:** DWs require continuous evolution, new datasets, and updated analytics tools.  
- **Focusing solely on ad hoc reporting:** Alert-driven reporting enhances near real-time operational decisions.  

**Key Insight:** Strong management support, sponsorship, and active user participation are **critical for DW success**.

---

### c. Scalability in Data Warehouses
Scalability refers to a DW’s ability to **handle data growth, concurrent users, and complex queries** without performance degradation.

- **Horizontal scaling:** Adding servers/nodes to distribute load.  
- **Vertical scaling:** Upgrading hardware (CPU, memory, storage).  
- **Data growth:** Includes current and historical archives.  
- **Query performance:** Ideally scales linearly with data volume.  

**Examples of Massive DWs:**
- Walmart: Hundreds of terabytes for sales/inventory analytics.  
- IBM: 50-terabyte benchmark warehouse.  
- U.S. Department of Defense: 5-petabyte medical repository.  
- CNN: Petabyte-scale media warehouse.  
- AT&T: 26-terabyte warehouse for fraud detection and large-scale analytics.  

**Techniques for scalability:** Parallel processing, indexing and partitioning, distributed storage across multiple locations.

---

### d. Case Study: Michigan State Government EDW

**Context:** Michigan’s Department of Technology, Management & Budget (DTMB) created a statewide enterprise DW to enhance service delivery, resource optimization, and decision-making.

**Scope:**
- ~10,000 users across 5 departments, 20 agencies, 100+ bureaus.  
- Integrated data from DHS, Community Health, and other agencies.

**Financial Impact:**
- $1 million/day in benefits.  
- DHS realized $200 million annually, plus $75 million for Human Services.  
- Savings include program integrity, cost avoidance, operational efficiency, and improper payment recovery.

**Operational Impact:**
- Supported strategic and tactical decisions across multiple agencies.  
- 15:1 cost-effectiveness ratio over 10 years.  
- Enterprise-wide sharing enabled innovative cross-agency analytics.

**Lessons Learned:**
- Enterprise-wide integration maximizes impact.  
- Centralized data removes barriers between agencies.  
- DWs provide measurable, monetary benefits.  
- User participation ensures relevance and usability.

---

### e. Human and Organizational Factors
- Strong executive sponsorship and a project champion are essential.  
- User engagement in **data and access modeling** ensures DW meets actual analytical needs.  
- DW teams require **technical expertise** in ETL, databases, and analytics tools.

---

### f. Offshore Development Considerations
- DW and BI support may involve **offshore resources**, growing at 20–25% annually.  
- **Risks:** Cultural differences, security, and data privacy must be carefully managed.

---

### g. Summary: DW Implementation Best Practices
- Align sponsorship, project champion, and technical leadership.  
- Set realistic expectations in selling and delivery phases.  
- Avoid political pitfalls; frame DW as **decision-enabling**.  
- Focus on relevant, curated, and clean data.  
- Design DWs for scalability (horizontal and vertical).  
- Incorporate alert-driven analytics, not just periodic reporting.  
- Treat DW as a **continuous, evolving system**.  
- Ensure user participation in data and access modeling.  
- Plan for integration of external and unstructured data where valuable.  
- Evaluate offshore resources cautiously, considering security and cultural factors.

## 3.8 Real-Time Data Warehousing (RDW / Active Data Warehousing, ADW)

### a. Definition and Concept
Real-Time Data Warehousing (RDW), also called Active Data Warehousing (ADW), is an evolution of traditional enterprise data warehouses (EDWs) designed to support **operational, tactical, and strategic decision-making** in near-real-time.

**Key Characteristics:**
- Data is loaded and made available immediately as it is generated, rather than in daily or weekly batch updates.  
- Enables decision-making at the point of action for employees, managers, and customers.  
- Integrates operational BI with traditional analytics, moving from historical reporting to influencing real-time outcomes.

**Evolution of Data Sophistication:**
- **Reporting:** “What happened?” – batch/ad hoc reports.  
- **Analyzing:** “Why did it happen?” – enhanced ad hoc analysis.  
- **Predicting:** “What will happen?” – predictive and analytical modeling.  
- **Operationalizing:** “What is happening now?” – continuous updates and time-sensitive queries.  
- **Activating:** “Make it happen!” – event-based triggers and enterprise decision management.  
*(Adapted from Figure 3.12: Enterprise Decision Evolution)*

---

### b. Drivers for Real-Time Data Warehousing
Organizations adopt RDWs to meet business requirements beyond the capabilities of batch-oriented DWs:

- **Need for fast decision-making:** Batch ETL is too slow for operational decisions.  
- **Operational integration:** BI tightly coupled with operational processes (e.g., customer service, supply chain).  
- **High-frequency data changes:** Incremental real-time updates required for accuracy.  
- **Limitations of traditional architectures:** Hub-and-spoke and multiple ODS systems increase complexity.  
- **E-commerce and digital business:** Instant insights for pricing, promotions, fraud detection, and competitive advantage.

**Examples:**
- **Overstock.com:** Real-time DW connections for timely decisions.  
- **Egg plc:** Customer Data Warehouse (CDW) refreshed near real-time; 10 TB, 16-CPU system enabled hundreds of marketing campaigns and faster customer-specific decisions.

---

### c. Benefits of RDW
- **Operational and strategic support:** Tactical and strategic decision-making in one system.  
- **Reduced latency:** Data availability shifts from hours/days to minutes/seconds.  
- **Event-driven analytics:** Supports proactive alerts and automated responses.  
- **Improved customer service:** Real-time data enables personalized interactions.  
- **Reduction/elimination of ODS:** Direct, continuous loading into DW.  
- **Enterprise-wide decision consistency:** Centralized, up-to-date data reduces conflicting reports.

**Example – Egg plc:**
- Processes 2.5 million daily transactions.  
- Hundreds of marketing campaigns executed within minutes.  
- Enables faster, data-driven decisions at customer and customer-segment level.

---

### d. Key Differences Between Traditional DW and RDW
| Feature | Traditional DW | Real-Time DW (RDW / ADW) |
|---------|----------------|---------------------------|
| Data Update Frequency | Batch (daily/weekly) | Continuous, near real-time |
| Primary Focus | Strategic and tactical reporting | Operational, tactical, strategic decisions |
| Decision Latency | Hours to days | Seconds to minutes |
| Integration with Operations | Minimal | Directly integrated with front-line operations |
| System Complexity | Moderate | High (concurrency management, event triggers, mini-batch ETL) |
| Use Cases | Trend analysis, forecasting | Customer service, marketing, fraud detection, SCM optimization |

---

### e. RDW Implementation Considerations
Implementing an RDW introduces new **architectural and operational challenges**:

- **Architecture & Data Modeling:** Real-time ingestion, concurrent queries, rapid aggregation.  
- **Physical Database Design:** Optimized for simultaneous high-speed reads/writes.  
- **Storage & Scalability:** Accommodate fast-growing volumes without performance degradation.  
- **Version Consistency:** Multiple “truths” may arise depending on access timing.  
- **ETL Redesign:** Replace nightly ETL with continuous or mini-batch feeds.  
- **Alert and Event Management:** Proactive monitoring and intelligent triggers.  
- **Selective Real-Time Updates:** Not all data require continuous refresh (e.g., historical data).

**Example – Teradata Active EDW:**
- **Active Load:** Mini-batch/trickle feeds in minutes or seconds.  
- **Active Access:** Operational decision support within 5 seconds.  
- **Active Events:** Rule-based automated actions.  
- **Active Workload Management:** Dynamic resource allocation.  
- **Active Enterprise Integration:** Alignment with enterprise architecture for intelligent decisioning.  
- **Active Availability:** 24/7 business continuity.

---

### f. Business Examples of Real-Time DW
| Organization | Industry | RDW Use Case | Outcome |
|-------------|---------|--------------|--------|
| Egg plc | Online banking | Near-real-time customer CDW for marketing & service | Faster customer-specific decisions; hundreds of campaigns in minutes |
| Continental Airlines | Aviation | Flight management dashboard | Real-time identification of flight network issues |
| UPS | Logistics | Package-flow optimization | Reduced 100M delivery miles; saved 14M gallons fuel/year |

---

### g. Challenges and Best Practices
**Challenges:**
- Complex architecture, modeling, and scaling.  
- Maintaining consistent “truth” across real-time updates.  
- High infrastructure costs and IT demands.  
- Risk of over-updating unnecessary data.

**Best Practices:**
- Plan with focus on **quality assurance and data governance**.  
- Define which data need **real-time vs. batch updates**.  
- Ensure alert systems complement ad hoc/periodic reporting.  
- Align RDW implementation with operational business needs.

## 3.9 Data Warehouse Administration, Security Issues, and Future Trends

### a. Data Warehouse Administration
A **Data Warehouse Administrator (DWA)** is responsible for the operational efficiency, performance, and security of a data warehouse. This role extends beyond a traditional Database Administrator (DBA) due to the **scale, complexity, and business-critical nature** of enterprise data warehouses.

**Key Responsibilities:**
- Monitor warehouse performance and efficiency.  
- Maintain data integrity and consistency.  
- Ensure availability and flexibility for future enhancements.  
- Support BI/DSS systems that facilitate managerial decision-making.  

**Skills Required for a DWA:**

- **Technical Expertise:**  
  - Knowledge of high-performance hardware, software, and networking.  
  - Understanding of database management, storage, and indexing strategies.  

- **Business Insight:**  
  - Familiarity with organizational decision-making processes.  
  - Ability to align warehouse design with business objectives.  

- **Communication Skills:**  
  - Coordinate with stakeholders and users for reporting, data access, and system changes.  

*(Reference: Benander et al., 2000)*

---

### b. Security and Privacy Concerns
Data warehouses store **sensitive and regulated information**, making security and privacy essential.

**Four Main Areas of Data Warehouse Security:**

1. **Corporate Policies and Procedures:**  
   - Executive-level involvement in creating and communicating security policies.  

2. **Logical Security Controls:**  
   - User authentication, role-based access, encryption, and access control.  

3. **Physical Security:**  
   - Restrict access to data centers and storage infrastructure.  

4. **Internal Control and Auditing:**  
   - Regular reviews and audits to ensure compliance with security and privacy standards.

**Example Technology – Ambeo (Embarcadero Technologies):**
- Monitors real-time data access, usage, and privacy.  
- Tracks user activity: who accessed what data and when.  
- Detects potential security breaches with minimal system overhead.  

**Special Consideration:** Mobile device access should generally be **read-only** to prevent misuse or breaches.

---

### c. Future Trends in Data Warehousing
Data warehousing is evolving under the influence of **big data, cloud computing, and advanced analytics**.

**Key Drivers:**

- **Data Sourcing:**  
  - Web, social media, and other big data sources.  
  - Open-source analytics tools for cost-effective solutions.  
  - SaaS and cloud-based DW for scalable access.  

- **Infrastructure Enhancements:**  
  - **Columnar Databases:** Faster queries and better compression.  
  - **Real-Time DW:** Frequent refresh of operational data for near-instant insights.  
  - **DW Appliances:** Pre-optimized systems for performance and simplified management.  
  - **In-Memory Storage:** Accelerates high-speed analytics.  
  - **In-Database Analytics:** Runs algorithms close to the data to improve efficiency.  

- **Data Management and Analytics:**  
  - **Master Data Management (MDM):** Ensures data quality, integration, and regulatory compliance.  
  - **Advanced Analytics:** Integrates predictive modeling, AI/ML, and statistical analysis.  
  - **Next-Generation DBMS:** Innovations in query optimization, storage, and analytics integration.

**Benefits of These Advances:**
- Accelerated decision-making.  
- Reduced total cost of ownership (TCO).  
- Ability to manage large, complex, and diverse datasets.  
- Support for real-time operational and strategic decisions.

## Key Terms in Data Warehousing

- **Active Data Warehousing (ADW):** A data warehouse that supports both strategic and operational/tactical decision-making by updating and delivering data in near–real-time.

- **Cube:** A multidimensional structure that allows fast analysis of data across multiple dimensions (e.g., time, geography, product).

- **Data Integration:** The process of combining data from different sources into a unified view, often for use in a data warehouse.

- **Data Mart:** A subset of a data warehouse, focused on a specific business area or department.

- **Data Warehouse (DW):** A central repository of integrated data from multiple sources designed to support decision-making, reporting, and analytics.

- **Data Warehouse Administrator (DWA):** A professional responsible for managing, maintaining, and ensuring the performance, security, and availability of a data warehouse.

- **Dependent Data Mart:** A data mart created from a central enterprise data warehouse.

- **Dimensional Modeling:** A design technique for data warehouses that organizes data into fact and dimension tables for efficient querying and analysis.

- **Dimension Table:** A table in a data warehouse that contains descriptive attributes (dimensions) related to facts (e.g., customer, product, time).

- **Drill Down:** The process of navigating from summary-level data to more detailed data.

- **Enterprise Application Integration (EAI):** Software and processes that enable different enterprise applications to communicate and share data seamlessly.

- **Enterprise Data Warehouse (EDW):** A centralized data warehouse that provides integrated data across the entire organization.

- **Enterprise Information Integration (EII):** Tools and processes that provide a virtual integrated view of data from multiple sources without physically moving it.

- **Extraction, Transformation, and Load (ETL):** The process of extracting data from source systems, transforming it into the desired format, and loading it into a data warehouse.

- **Independent Data Mart:** A data mart created directly from operational systems, without using a central data warehouse.

- **Metadata:** Data about data; information describing the structure, content, and meaning of data in a data warehouse.

- **OLTP (Online Transaction Processing):** Systems designed for day-to-day transaction processing, such as order entry or banking.

- **Oper Mart:** Another term for an operational data mart, focusing on short-term, tactical decision support.

- **Operational Data Store (ODS):** A database designed to integrate data from multiple sources for operational reporting, often serving as an intermediate step before loading a data warehouse.

- **Real-Time Data Warehousing (RDW):** A data warehouse that updates data continuously or near–real-time, supporting immediate decision-making.

- **Snowflake Schema:** A data warehouse schema where dimension tables are normalized into multiple related tables, forming a “snowflake” structure.

- **Star Schema:** A simple data warehouse schema with a central fact table connected directly to denormalized dimension tables, forming a “star” pattern.
