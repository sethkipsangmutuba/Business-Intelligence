## 2.1 Opening Vignette: Decision Modeling at HP Using Spreadsheets

HP, a leading manufacturer of computers, printers, and industrial products, manages a vast and complex product line that generates numerous decision challenges. Olavson and Fry (2008) studied spreadsheet models used at HP and highlighted insights from successes and failures in creating decision-support tools.

---

### Definition of a Tool
A **tool** is a reusable analytical solution designed for non-technical end users to assist in solving repeated business problems.

---

### Phases in Developing a Decision-Support Model at HP

#### 1. Problem Framing
Key questions to consider before building a model:

- Will analytics actually solve the problem?  
- Can an existing solution be leveraged?  
- Is a new tool necessary?  

**Insights:**

- Not all problems are analytic; some require addressing organizational issues such as misaligned incentives or unclear authority.  
- Existing tools or decision guidelines may suffice, avoiding unnecessary development costs.  
- Simple decision rules or guidelines can sometimes replace complex models, simplifying training and adoption.

---

#### 2. Tool Design and Development
**Five key guidelines for successful model development:**

1. **Develop a prototype quickly** – Rapid iteration improves understanding and identifies issues early.  
2. **Test designs and features early** – Early testing ensures relevance and usability.  
3. **Gather feedback from users** – Continuous user input improves alignment with real-world needs.  
4. **Avoid overbuilding; stop when the tool is “good enough”** – Focus on insight, not black-box complexity.  
5. **Users should understand the problem and explore alternative scenarios** – Tools should support informed decision making, not just provide answers.  

**Additional Design Considerations:**

- **Remove unneeded complexity:** Simplify the tool to reduce training, errors, and recalibration needs.  
- **Partner with end users:** Collaboration improves problem understanding and tool adoption; users gain analytical knowledge and can refine the tool.  
- **Develop an Operations Research (OR) champion:** Champions promote tool adoption, support colleagues, and act as local experts.

---

#### 3. Handoff and Implementation
Key considerations when delivering the final tool:

- Identify who will use it and who owns the decisions it supports.  
- Define roles and responsibilities for maintenance and enhancement.  
- Align timing and integration with existing processes.  
- Assess impact on business performance.  
- Adjust metrics and incentives as necessary.

---

### Key Lessons for Decision Support

- Fully understand the problem before building a model.  
- Recognize that a model may not always be necessary.  
- Explore reuse of existing tools before creating new ones.  
- Focus on providing insight, not just additional numbers.  
- Plan implementation alongside model development to ensure successful adoption.

## 2.2 Decision Making: Introduction and Definitions

This section explores the practice of decision making, relevant theories and models, and traits of effective decision makers. Understanding these dimensions aids in identifying suitable decision-support tools to enhance managerial decisions.

---

### Characteristics of Decision Making

Decision making involves multiple challenges:

- **Groupthink:** Accepting solutions without critical evaluation can lead to poor outcomes.  
- **What-if scenarios:** Decision makers often simulate alternatives to evaluate possible outcomes.  
- **Experimentation:** Testing solutions in real systems is risky, costly, and often limited to single scenarios.  
- **Changing environment:** Dynamic conditions (e.g., seasonal variations) can invalidate assumptions and impose time pressure.  

**Information challenges:**

- **Insufficient information:** May prevent intelligent choices.  
- **Information overload:** Too much data can hinder decision quality.  
- **Cost and timing:** Collecting and analyzing information is expensive and time-consuming; knowing when to act is critical.  

**Role of DSS:** Decision Support Systems provide structured, rational approaches for evaluating alternatives efficiently and mitigating these challenges.

---

### Working Definition of Decision Making

- **Decision making:** The process of choosing among two or more alternative actions to achieve one or more goals.  
- **Managerial decision making:** Encompasses the full management process, including planning, organizing, and controlling.  
- **Example:** Planning involves decisions about what, when, where, how, and by whom tasks should be executed.

---

### Decision-Making Disciplines

Decision making draws upon multiple disciplines:

- **Behavioral disciplines:** Anthropology, law, philosophy, political science, psychology, social psychology, sociology.  
- **Scientific disciplines:** Computer science, decision analysis, economics, engineering, natural sciences, management science/operations research, mathematics, statistics.  

**Focus:** DSS and Management Support Systems (MSS) prioritize decision effectiveness over computational efficiency.

---

### Decision Style and Decision Makers

**Decision style** reflects how individuals perceive problems, think, and react, shaped by cognitive responses, values, and beliefs. Decision processes are not strictly linear; steps may be skipped, repeated, or reordered.

**Common styles:**

- Heuristic vs. analytic  
- Autocratic vs. democratic  
- Consultative (individual or group)  

**Combination of traits:** Examples include analytic-autocratic or consultative-heuristic.

**System implications:**  

- DSS should accommodate multiple decision styles rather than enforce a rigid process.  
- Tools that support what-if analysis and goal-seeking enhance flexibility and user effectiveness.  
- Web-based interfaces and collaborative platforms facilitate group decision support.

**Decision makers:**

- **Individuals:** Often handle decisions at lower managerial levels or in small organizations, balancing multiple objectives.  
- **Groups:** Manage major decisions in medium to large organizations; conflicts in objectives, styles, and personalities increase complexity.  

**DSS support for groups:**  
Platforms such as Enterprise Information Systems (EIS), Group Support Systems (GSS), ERP/ERM, SCM, KMS, and CRM systems improve collaboration and decision quality.

## 2.3 Phases of the Decision-Making Process

A systematic approach to decision making enhances outcomes. Simon (1977) identified four major phases: **intelligence, design, choice, and implementation**. Monitoring is often integrated into the intelligence phase as applied to implementation, providing feedback.

---

### Overview of the Process

The decision-making process typically flows as:  

$$\text{Intelligence} \rightarrow \text{Design} \rightarrow \text{Choice} \rightarrow \text{Implementation}$$  

**Feedback loops** allow revisiting earlier phases when adjustments are necessary. Modeling is central, explaining why decision making often appears non-linear or iterative.

---

### Intelligence Phase

Focus: Scanning the environment to identify problems or opportunities.

**Activities:**

- Data collection and analysis  
- Problem identification and classification  
- Establishing problem ownership  

**Monitoring** of implemented decisions is also included, ensuring feedback informs subsequent intelligence efforts.

---

### Design Phase

Focus: Constructing a **model of the system** to simplify reality and define variable relationships.

**Activities:**

- Formulating assumptions and simplifications  
- Identifying alternative solutions  
- Validating the model  
- Establishing criteria for evaluating alternatives  

**Note:** Model building and alternative identification are often **iterative**.

---

### Choice Phase

Focus: Selecting a solution from the model (may not directly solve the real-world problem).

**Activities:**

- Testing solution viability  
- Performing sensitivity analysis  

**Outcomes:** Success leads to implementation; failure may require revisiting the design or intelligence phase.

---

### Implementation Phase

Focus: Executing the chosen solution in the real environment.

- Effective implementation solves the actual problem.  
- If the solution fails, feedback loops return the process to earlier phases.

---

### Web Impacts on Decision Making

| Phase       | Web Impacts | Example Impacts |
|------------|-------------|----------------|
| Intelligence | Access to internal/external data, analytics methods, intelligent agents, collaboration tools (GSS, KMS) | Identification of e-commerce and IT opportunities |
| Design      | Access to data, models, OLAP, data warehouses, collaboration | Sharing and developing Web infrastructure solutions |
| Choice      | Tools for evaluating alternatives and setting criteria | DSS tools help determine Web, intranet, and extranet design decisions |
| Implementation | Web-based collaboration tools and KMS | Performance monitoring of e-commerce, intranet, extranet, and Internet systems |

---

### Alternative Decision-Making Models

Structured methods like **Kepner-Tregoe** align with Simon’s four-phase model, offering practical frameworks for organizations.

---

### Key Points

- Systematic approaches ensure decisions are **rational, informed, and adaptable**.  
- **Feedback loops** are critical; solutions often require iterative refinement.  
- Technology, especially **Web-based tools**, enhances all phases of the decision-making process.

## 2.4 Decision Making: The Intelligence Phase

The **intelligence phase** involves scanning the organizational and external environment to identify problems or opportunities. This may occur continuously or intermittently and often includes monitoring the outcomes of previous decisions.

---

### Problem (or Opportunity) Identification

- Begins with **defining organizational goals and objectives**, e.g., inventory management, staffing, or online presence.  
- **Problems arise** when there is dissatisfaction with the status quo, i.e., a gap between expectations and reality.  

**Key Steps:**

1. Determining whether a problem exists  
2. Identifying **symptoms** vs. the **actual problem**  
3. Assessing the **magnitude and scope** of the problem  
4. Explicitly defining the problem  

**Example:** In the *“Elevators Go Faster”* case, perceived waiting time—not actual waiting time—was the real problem. Installing mirrors reduced complaints effectively and inexpensively, highlighting the importance of identifying the **correct problem**.

---

### Data Collection and Estimation Issues

Challenges include:

- Data may be **missing, inaccurate, or subjective**  
- Collection can be **expensive or time-consuming**  
- Data may be **qualitative** or **overwhelming in volume**  
- Outcomes may occur **over time**, requiring **present-value calculations**  
- **Future trends** may not follow historical patterns  

Accurate problem identification depends on distinguishing **real problems** from mere **symptoms**.

---

### Problem Classification

- Classifying problems determines their **structuredness**, from **structured (programmed)** to **unstructured (unprogrammed)**.  
- Classification can guide the use of **standardized solution approaches**.

---

### Problem Decomposition

- Complex problems are broken into **subproblems** for easier resolution.  
- **Structured subproblems** may exist within otherwise unstructured issues.  
- Decomposition facilitates **communication** among decision makers and supports methods like the **Analytical Hierarchy Process (AHP)**.

---

### Problem Ownership

- Assigns **responsibility** for solving a problem to an individual or group.  
- Ensures **accountability** and clarifies which problems are actionable.  

**Example:** High interest rates cannot be controlled by a company; the actionable problem is how to operate effectively in a **high–interest-rate environment**.

---

### Conclusion of the Intelligence Phase

- Results in a **formal problem statement** specifying the **real problem, its scope, and ownership**.  
- Provides the **foundation** for the next phase: **designing models and exploring alternatives**.

## 2.5 The Design Phase of Decision Making

The **design phase** focuses on creating and analyzing possible courses of action. It emphasizes understanding the problem, exploring alternatives, and testing feasibility rather than immediately selecting a solution. Central to this phase is the development of **models**, which allow decision makers to experiment, simulate, and analyze outcomes without affecting real-world operations.

---

### Models in Decision Support Systems (DSS)

A **model** is a simplified representation of reality that abstracts away unnecessary details to focus on relevant aspects of a problem.

**Why Models Are Used:**

- **Safe experimentation:** Test alternatives without disrupting operations.  
- **Time compression:** Simulate long-term operations in minutes.  
- **Cost efficiency:** Cheaper than real-world testing.  
- **Risk estimation:** Evaluate potential outcomes and uncertainties.  
- **Handling complexity:** Analyze multiple alternatives simultaneously.  
- **Learning and training:** Enhance understanding of complex systems.

---

### Types of Models

#### Mathematical (Quantitative) Models

- Represent relationships **numerically** (e.g., linear programming, simulation, forecasting).  
- **Variables:**
  - **Decision variables:** Alternatives to choose from (e.g., production quantity).  
  - **Result variables:** Outcomes reflecting goals (e.g., profit, revenue).  
  - **Uncontrollable variables:** Environmental factors (e.g., economic conditions).  

**Principle of Choice:** Criterion used to evaluate solutions (e.g., maximize profit, minimize cost).  
**Criteria vs Constraints:**  
- **Criterion:** Expresses a preference.  
- **Constraint:** Sets a hard limit (e.g., cannot travel beyond 100 km).

---

###  Normative vs Descriptive Models

#### Normative Models

- Aim to identify the **best possible alternative** (optimization).  
- Assume **rational decision makers** who:  
  - Maximize goals  
  - Know all alternatives and outcomes  
  - Can rank outcomes by preference  

**Optimization Methods:**

- Maximize goal attainment from given resources  
- Maximize ratio of goal attainment to cost  
- Minimize resources while achieving acceptable goals  

**Suboptimization:**  
- Occurs when only part of a system is optimized rather than the whole.  
- **Example:** Marketing drives high demand but production cannot meet it.  

#### Descriptive Models

- Represent reality as perceived, not necessarily the “best” solution.  
- Examples: simulation, cognitive maps, narrative models  
- Useful for evaluating consequences and training managers.

---

###  Satisficing and Bounded Rationality

- **Satisficing:** Choosing a solution that meets **minimum acceptable criteria**, not necessarily optimal.  
- **Bounded rationality:** Decision makers process **limited information**, simplifying problems.  
- **Implication:** Intuition is often relied upon in unstructured or urgent decisions.

---

###  Developing Alternatives

- Generating alternatives is **creative and often manual**, though models can automate some.  
- Considerations:
  - Cost and availability of information  
  - Expertise required  
  - Risk of information overload  
- Sequence: Define criteria first, then search for alternatives.

---

###  Measuring Outcomes

- Outcomes evaluated based on **goal attainment**:
  - **Quantitative:** Profit, cost, revenue  
  - **Qualitative:** Customer satisfaction, loyalty, ratings  
- Multi-criteria analysis (e.g., AHP) is used when goals conflict.

---

### Risk and Uncertainty

- Decision environments:
  - **Certainty:** Outcomes known  
  - **Risk:** Outcomes uncertain, probabilities known  
  - **Uncertainty:** Outcomes and probabilities unknown  
- Decision makers must **estimate risk** and assess whether alternatives are acceptable.

---

### Scenarios

- **Scenario:** Narrative describing operating environment and decision context  
- Used for **what-if analysis, simulations, and planning**  
- Types: worst-case, best-case, most likely, average  
- Helps identify opportunities, validate assumptions, and explore system behavior.

---

### Errors in Decision Making

Common errors include:

- Using incomplete or inaccurate data  
- Overconfidence or illusion of control  
- Relying excessively on intuition  
- Improperly defining criteria and constraints  

**Mitigation:** Model validation and careful data collection reduce errors.


## 2.6 Decision Making: The Choice Phase

**Definition:**  
The **choice phase** is the stage where an actual decision is made, and a commitment to act is established. It involves selecting one alternative from those generated during the design phase.

---

### Key Points

- **Blurry boundary with the design phase:** New alternatives may be generated while evaluating options.  
- **Solving the model ≠ solving the problem:** The problem is only truly resolved when the chosen solution is successfully implemented.

---

### Approaches to Searching for Solutions

- **Analytical techniques:** Solve formulas derived from models.  
- **Algorithms:** Step-by-step procedures to systematically identify solutions.  
- **Heuristics:** Experience-based rules of thumb or shortcuts.  
- **Blind searches:** Random or exploratory approaches, applied logically when necessary.

---

### Evaluating Alternatives

- **Multiple goals:** Alternatives are evaluated to balance conflicting objectives.  
- **Sensitivity analysis:** Tests how small changes in parameters affect outcomes.  
- **What-if analysis:** Explores effects of major parameter changes.  
- **Goal seeking:** Determines decision variable values to achieve a specific target or objective.

---

### Importance

- The choice phase represents the **moment of commitment**; without making a choice, the decision remains theoretical.  
- Ensures that the **best available alternative** is acted upon, grounded in modeling and evaluation.

---

## 2.7 Decision Making: The Implementation Phase

**Definition:**  
The **implementation phase** is where a recommended solution is put into action, effectively initiating change. It extends beyond system installation to ensuring that the decision works in practice.

---

### Key Points

**Implementation = Managing Change**  
- Introducing a new solution involves inherent risk and complexity.  
- Change management is essential to address user expectations, resistance, and organizational culture.  
- Many technology projects (e.g., BPR, knowledge management) fail due to inadequate change management.

**Role of Management Support and Training**  
- Strong support from top management increases the probability of successful implementation.  
- User training equips employees to use new processes or systems effectively.

**Project Management**  
- Implementation requires careful planning and control.  
- Knowledge of project management principles is critical for scheduling, resource allocation, and risk mitigation.  
- Professional certifications (e.g., PMP) highlight the value of structured project management.

**Feedback and Learning**  
- Collecting data during implementation allows assessment of outcomes and informs future decisions.  
- Analytics can verify whether the initial problem was correctly identified and whether the solution is effective.  
- Feedback loops are crucial for continuous improvement and public policy evaluation.

**Computer Support**  
- DSS and business analytics support implementation by monitoring performance, automating processes, and providing timely information for adjustments.

---

**Objective:** Generate alternatives and evaluate potential outcomes.

- **Standard DSS models:** Financial, forecasting, and process models.

**Feedback Loop:** DSS and BI technologies report outcomes, feeding insights back into the intelligence phase for continuous improvement.

## 2.9 Decision Support Systems: Capabilities

A **Decision Support System (DSS)** is a computer-based system designed to assist managerial decision making in **semi-structured and unstructured situations**. It does **not replace human judgment** but integrates human insight with computerized data, models, and analytical tools to improve decision quality.

---

### Key Points About DSS

**Problem Focused**  
- Designed to solve a specific problem or set of related problems.  
- Distinct from Business Intelligence (BI) systems, which mainly monitor and report data.

**Interactive and Flexible**  
- Supports iterative analysis and adapts to similar problems.  
- Offers user-friendly interfaces, often web or mobile-based.

**Decision Phases Supported**  
- Assists managers across **intelligence, design, choice, and implementation** phases.

**Support Across Levels and Groups**  
- Can be used by individuals, teams, or groups at all managerial levels.  
- Supports interdependent or sequential decisions.

**Use of Models**  
- Incorporates **quantitative or qualitative models** to simulate and analyze alternatives.  
- Enables experimentation with different strategies before implementation.

## 2.10 DSS Classifications

Decision Support Systems (DSS) can be classified in multiple ways. The type of DSS determines its **design, operation, and implementation**, though many DSS combine characteristics of more than one category.

---

### AIS SIGDSS Classification (Power, 2002)

**1. Communications-Driven and Group DSS (GSS)**  
- Supports groups in **collaborative work and decision making**.  
- Examples: design collaboration, meetings, supply chain management, KMS for collaborative communities.

**2. Data-Driven DSS**  
- Focuses on **data access, processing, and presentation**.  
- Minimal use of models; relies on **data warehouses, OLAP, and reporting tools**.

**3. Document-Driven DSS**  
- Uses **text, documents, and multimedia** to support decision making.  
- Minimal modeling; relies on **knowledge storage and retrieval**.  
- Example: U.S. Army Defense Ammunitions Center system.

**4. Knowledge-Driven DSS / Data Mining / Expert Systems**  
- Uses **AI, rules, expert systems (ES), and neural networks**.  
- Supports **rapid decision making** in situations requiring intelligent automation.

**5. Model-Driven DSS**  
- Focuses on **optimization, simulation, and model analysis**.  
- Used for **large-scale decision problems and what-if analysis**.  
- Example: Excel-based linear programming or financial forecasting models.

**6. Compound (Hybrid) DSS**  
- Combines **two or more DSS types**.  
- Example: WolframAlpha combines data, models, documents, and algorithms.

---

### Other DSS Categories

**Institutional DSS**  
- Supports **recurring decisions** over time.  
- Evolves and improves through repeated use.  
- Example: Portfolio management systems in banks.

**Ad Hoc DSS**  
- Supports **nonrecurring or one-time decisions** (e.g., strategic planning).  
- May evolve into institutional DSS if reused or applied elsewhere.

---

### Custom-Made vs Ready-Made DSS

**Custom-Made DSS**  
- Built for a **specific organization or user**.

**Ready-Made DSS**  
- Generic DSS adaptable by **multiple organizations**.  
- Often includes templates, OLAP, and web-based interfaces.  
- Examples: Cognos, MicroStrategy, Teradata DSS solutions.

> **Note:** Many modern DSS are **hybrid**, combining features from multiple types (e.g., data + models + knowledge + collaboration).

---

### Key Insights

- DSS can support **individuals, groups, and organizations** in different decision contexts.  
- **Classification influences design and usage**.  
- Modern DSS often integrate **data, models, knowledge, and collaborative features**.

## 2.11 Components of DSS

A Decision Support System (DSS) typically consists of **four main subsystems**:

---

### 1. Data Management Subsystem

**Purpose:** Stores and manages relevant data for decision making.

**Components:**

- **DSS Database:** Stores data specific to the DSS application.  
- **Database Management System (DBMS):** Manages data storage, retrieval, and updates.  
- **Data Directory:** Lists data sources and metadata.  
- **Query Facility:** Enables users to retrieve and analyze data.  
- **Integration:** Can link with corporate data warehouses, ERP, POS, legacy systems, Web/internet sources.

**Example:** Station Casinos used **500+ data sources** to build a DSS for **customer segmentation and marketing optimization**.

---

### 2. Model Management Subsystem

**Purpose:** Provides analytical capabilities for problem-solving.

**Components:**

- **Model Base:** Collection of models (financial, statistical, management science, optimization, simulation).  
- **Model Base Management System (MBMS):** Software for managing models.  
- **Modeling Language:** Builds or customizes models.  
- **Model Directory:** Lists available models and their functions.  
- **Model Execution & Integration:** Runs models and integrates results with other DSS components.

**Usage:** Ranges from simple (**Excel/spreadsheets**) to complex (**Java, C++, .NET, simulation/statistics packages**).  

**Example:** OneNet’s **SNAP-DSS** used Excel + VBA for **telecommunications pricing analysis**.

---

### 3. User Interface Subsystem

**Purpose:** Facilitates interaction between users and DSS.

**Features:**

- GUIs: Web browsers, dashboards, portals, spreadsheets.  
- Input methods: Text, natural language, voice commands, mobile apps, SMS, RFID/IoT sensors.  
- Enables real-time analysis, intuitive understanding, and scenario exploration.

**Examples:** Amazon mobile app (product recognition & comparison), Google Now, Apple Siri, SMS-based queries.

---

### 4. Knowledge-Based Management Subsystem (Optional)

**Purpose:** Enhances decision-making by adding **intelligence and learning**.

**Integration:** Augments other subsystems or acts independently; often connected to **Knowledge Management Systems (KMS)**.  

**Techniques:** AI, machine learning, natural-language processing, evidence-based reasoning.

**Example:** **IBM Watson** supports **evidence-based medical decisions**, analyzing vast structured and unstructured datasets.

---

### Key Insights

- **Core components:** DBMS, MBMS, and user interface.  
- **Knowledge-based subsystem:** Optional, adds significant intelligence.  
- Modern DSS are increasingly:
  - Web-based and mobile-accessible  
  - Integrated with ERP, CRM, SCM, BPM, GIS  
  - Embedded in real-time business processes  
  - Supporting collaboration and group decision-making  
- **Emerging technologies:** Multi-touch, gesture recognition, EEG interfaces, wearable devices, IoT sensors, enhancing accessibility and interactivity.

---

### Review Questions – Brief Answers

**What is a cognitive system?**  
A system using AI and analytics to simulate human thinking and support real-time decision making.

**What is evidence-based decision making?**  
Decisions guided by data, analytics, and verifiable knowledge rather than intuition.

**Role of Watson in DSS:**  
Knowledge-based DSS providing real-time, evidence-based recommendations in healthcare and insurance using AI and large-scale data analysis.

**Does Watson replace human decision making?**  
No; it **augments** human decision making, offering insights and recommendations while humans retain final judgment.

# Highlights – Key Points

---

## 1. Decision Making and Management

- Managerial decision making is central to **all management processes**.  
- **Human decision styles** must be considered when designing systems.  
- Decision support can target **individuals, groups, and organizations**.

---

## 2. Problem Solving

- Decision making involves **solving problems** and **evaluating opportunities**.  
- **Models** are simplified representations of reality that help **analyze alternatives**.

---

## 3. Four Phases of Decision Making

### Intelligence Phase
- Identify, classify, and decompose problems or opportunities.  
- Establish **problem ownership**.

### Design Phase
- Build models, generate alternatives, predict results, and create decision methodology.

### Choice Phase
- Compare alternatives and select the **best or satisfactory solution**.  
- Techniques include **optimization, search, and what-if analysis**.

### Implementation Phase
- Apply solutions considering **multiple goals** and **sensitivity analysis**.  
- **Satisficing:** Accepting a satisfactory solution due to **bounded rationality**.

---

## 4. Role of Computer Systems

- Support **all phases of decision making**.  
- Automate tasks and apply **artificial intelligence**.

---

## 5. Decision Support Systems (DSS)

- Support **complex managerial problems** that other computerized systems cannot solve.  
- **User-oriented:** Uses data, models, and user input.  

**DSS vs. BI systems:**
- **DSS:** Solve specific problems.  
- **BI:** Monitor situations and provide analytical tools after problems are identified.  

- Supports **all managerial levels** and both **individual and group decisions**.  
- Improves **decision effectiveness, communication, management control, objectivity, and efficiency**.

---

## 6. DSS Classifications (AIS SIGDSS)

- **Communications-driven / Group DSS (GSS)**  
- **Data-driven DSS**  
- **Document-driven DSS**  
- **Knowledge-driven DSS / Data mining / Expert Systems**  
- **Model-driven DSS**  
- **Compound DSS** (hybrid of two or more types)

---

## 7. Other DSS Classifications

- **Purpose:** Institutional vs. Ad hoc DSS  
- **Organizational Level:** Personal, group, or organizational support  
- **Work Focus:** Individual DSS vs. GSS  
- **Development Type:** Custom-made vs. Ready-made DSS

---

## 8. DSS Components

### Data Management Subsystem
- DSS database, DBMS, data directory, query facility.  
- Integrates internal and external sources (**ERP, data warehouse, Web**, etc.).

### Model Management Subsystem
- Standard and custom models, managed via **MBMS**.  
- Built using programming, modeling, or **Web-based systems**.

### User Interface Subsystem
- Web browsers, dashboards, smartphones, and tablets.  
- Provides **intuitive, consistent interaction**.

### Knowledge-Based Management Subsystem (optional)
- Adds **intelligence and learning capabilities** (e.g., IBM Watson).

---

## 9. Recent Trends in DSS

- Integration with **data warehouses, OLAP, and data mining**.  
- **User interfaces** increasingly mobile, web-based, and intuitive.  
- DSS components are **integrated, embedded, and enhanced** by AI, sensors, and wearable devices.
